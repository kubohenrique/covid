# Panorama do COVID-19 no Brasil

O coronavírus é uma família de vírus que pode causar doenças em animais e seres humanos. Em dezembro de 2019, um novo coronavírus, chamado SARS-CoV-2, foi identificado em Wuhan, na China, como a causa de uma doença respiratória que ficou conhecida como COVID-19.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

A COVID-19 se espalhou rapidamente pelo mundo, sendo declarada uma pandemia pela Organização Mundial da Saúde (OMS) em março de 2020. A doença tem sintomas variados, que podem incluir febre, tosse, falta de ar, fadiga, dor de cabeça, perda de paladar ou olfato, entre outros. Em casos mais graves, pode levar à pneumonia, síndrome respiratória aguda grave, falência de órgãos e até mesmo à morte, especialmente em pessoas idosas ou com condições médicas pré-existentes.

Para conter a propagação do vírus, muitos países implementaram medidas de saúde pública, como o uso de máscaras faciais, distanciamento social, restrições de viagem, fechamento de escolas e empresas, e campanhas de vacinação em massa.

A pandemia de COVID-19 teve um impacto significativo em todos os aspectos da vida, incluindo a saúde pública, economia, educação, viagens e interações sociais. Apesar dos esforços para controlar a disseminação do vírus, a pandemia continua a evoluir, com novas variantes do vírus surgindo e desafios contínuos na distribuição equitativa de vacinas.

É importante seguir as orientações das autoridades de saúde e continuar a tomar medidas preventivas para proteger a si mesmo e aos outros contra a COVID-19.

# COVID-19 no Brasil

No Brasil, a COVID-19 teve um impacto significativo desde o seu surgimento. O primeiro caso confirmado no país foi registrado em fevereiro de 2020, e desde então a doença se espalhou rapidamente por todas as regiões, afetando milhões de pessoas.

O país enfrentou desafios significativos na gestão da pandemia, incluindo subnotificação de casos, falta de leitos hospitalares e equipamentos de proteção, bem como dificuldades na implementação de medidas de distanciamento social devido a questões socioeconômicas e políticas.

A COVID-19 sobrecarregou o sistema de saúde brasileiro, especialmente em áreas mais pobres e remotas, onde o acesso a cuidados médicos de qualidade é limitado. O aumento nos casos também levou a um alto número de mortes, colocando pressão adicional sobre os serviços funerários e aumentando o sofrimento das famílias afetadas.

Além dos impactos na saúde, a pandemia teve um grande impacto na economia brasileira, com fechamento de empresas, aumento do desemprego e queda na atividade econômica. O governo implementou várias medidas de auxílio para mitigar esses efeitos, como o auxílio emergencial, que forneceu assistência financeira a milhões de brasileiros afetados pela crise.

A vacinação contra a COVID-19 começou no Brasil em janeiro de 2021, mas enfrentou desafios logísticos e de distribuição. Apesar dos esforços contínuos para vacinar a população, o país ainda enfrenta desafios na cobertura vacinal e na emergência de variantes do vírus que podem afetar a eficácia das vacinas.

Em resumo, a COVID-19 teve um impacto profundo no Brasil, afetando não apenas a saúde pública, mas também a economia, o sistema de saúde e a vida das pessoas em todo o país. A gestão da pandemia continua sendo uma prioridade, com esforços contínuos para conter a propagação do vírus e minimizar seus impactos.

## Desafios de juntar dados à epoca da pandemia

Analisar uma nova doença, como a COVID-19, apresenta diversos desafios devido à natureza complexa e em constante evolução do vírus, bem como às diferenças nos protocolos de coleta e relatório de dados entre governos e fontes não oficiais. Aqui estão alguns dos principais desafios:

1. **Falta de dados consistentes e confiáveis**: A coleta de dados sobre uma nova doença pode ser inconsistente e sujeita a erros, especialmente no início da pandemia quando os protocolos de teste e relatório podem variar entre países e regiões. Isso pode levar a discrepâncias nos números relatados de casos e mortes.

2. **Diferenças nos protocolos de teste e relatório**: Os critérios para realizar testes de COVID-19 e relatar casos e mortes podem variar entre governos federal, estaduais e fontes não oficiais. Isso pode levar a discrepâncias nos números relatados e dificultar a comparação entre diferentes fontes de dados.

3. **Interpretação e apresentação dos dados**: Mesmo quando os dados são consistentes, sua interpretação e apresentação podem variar entre governos e fontes não oficiais. Isso pode levar a diferentes conclusões sobre a gravidade da pandemia e as medidas necessárias para controlá-la.

4. **Política e viés**: As autoridades políticas podem ter incentivos para manipular ou distorcer os dados para atender a seus próprios interesses políticos. Isso pode levar à subnotificação de casos e mortes, bem como à minimização da gravidade da pandemia.

5. **Acesso limitado a dados**: Algumas fontes de dados, como dados hospitalares e registros de morte, podem não estar disponíveis publicamente ou podem ser controladas pelo governo, dificultando o acesso a informações precisas e atualizadas.

Em suma, a análise da COVID-19 sob a ótica dos diferentes dados fornecidos pelos governos federal, estaduais e fontes não oficiais é complexa e sujeita a diversos desafios. É importante abordar esses desafios com transparência, colaboração e rigor metodológico para garantir que as decisões políticas e de saúde pública sejam baseadas em evidências sólidas e confiáveis.

## Our World in Data (OWID)

A Our World in Data é uma plataforma online que fornece visualizações interativas e análises de dados sobre uma ampla variedade de tópicos globais, incluindo saúde, meio ambiente, economia, educação e muito mais. A plataforma foi fundada em 2011 pelo economista Max Roser e sua equipe da Universidade de Oxford.

Uma das características distintivas da Our World in Data é seu compromisso com a apresentação de dados confiáveis e baseados em evidências, com fontes claramente citadas e métodos de coleta de dados transparentes. Eles trabalham em estreita colaboração com instituições de pesquisa e organizações internacionais para reunir e analisar os dados mais atualizados disponíveis.

A plataforma oferece uma variedade de recursos, incluindo gráficos interativos, artigos detalhados, conjuntos de dados para download e ferramentas de visualização personalizadas. Esses recursos são projetados para ajudar os usuários a entender melhor os padrões e tendências globais, bem como a explorar questões específicas em profundidade.

Our World in Data é amplamente reconhecida como uma fonte confiável de informações sobre uma ampla gama de tópicos globais. Sua missão é capacitar as pessoas com dados para que possam tomar decisões informadas e contribuir para um mundo melhor.

* `new_cases_smoothed` - Novos casos confirmados de COVID-19 (suavizados em média de 7 dias). As contagens podem incluir casos prováveis, quando relatados.
* `total_deaths` -  Total de mortes atribuídas ao COVID-19. As contagens podem incluir mortes prováveis, quando relatadas.
* `new_deaths` -  Novas mortes atribuídas ao COVID-19. As contagens podem incluir mortes prováveis, quando relatadas. Em casos raros em que nossa fonte relata uma mudança diária negativa devido a uma correção de dados, essa métrica é definida como NA (não disponível).
* `new_deaths_smoothed` -  Novas mortes atribuídas ao COVID-19 (suavizadas em média de 7 dias). As contagens podem incluir mortes prováveis, quando relatadas.
* `total_cases_per_million` -  Número total de casos confirmados de COVID-19 por 1.000.000 de pessoas. As contagens podem incluir casos prováveis, quando relatados.
* `new_cases_per_million` -  Novos casos confirmados de COVID-19 por 1.000.000 de pessoas. As contagens podem incluir casos prováveis, quando relatados.
* `new_cases_smoothed_per_million` -  Novos casos confirmados de COVID-19 (suavizados em média de 7 dias) por 1.000.000 de pessoas. As contagens podem incluir casos prováveis, quando relatados.
* `total_deaths_per_million` -  Total de mortes atribuídas ao COVID-19 por 1.000.000 de pessoas. As contagens podem incluir mortes prováveis, quando relatadas.
* `new_deaths_per_million` -  Novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas. As contagens podem incluir mortes prováveis, quando relatadas.
* `new_deaths_smoothed_per_million` -  Novas mortes atribuídas ao COVID-19 (suavizadas em média de 7 dias) por 1.000.000 de pessoas. As contagens podem incluir mortes prováveis, quando relatadas.
* `reproduction_rate` -  Estimativa em tempo real da taxa efetiva de reprodução (R) do COVID-19.
* `icu_patients` - Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia.
* `icu_patients_per_million` -  Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia por 1.000.000 de pessoas.

* `hosp_patients_per_million` -  Número de pacientes com COVID-19 no hospital em um determinado dia por 1.000.000 de pessoas.
* `weekly_icu_admissions` -  Número de pacientes com COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma semana específica (data do relatório e nos 6 dias anteriores).
* `weekly_icu_admissions_per_million` - Número de pacientes com COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma semana específica por 1.000.000 de pessoas (data do relatório e nos 6 dias anteriores).

* `new_tests` - Novos testes para COVID-19 (calculados apenas para dias consecutivos).
* `total_tests_per_thousand` - Total de testes para COVID-19 por 1.000 pessoas.
* `new_tests_per_thousand` - Novos testes para COVID-19 por 1.000 pessoas.

Isso produz uma série completa de números diários, que é então calculada em uma janela móvel de 7 dias.

* `tests_per_case` - Testes realizados por novo caso confirmado de COVID-19, apresentado como uma média móvel de 7 dias (isso é o inverso da taxa de positividade).

Todas as métricas relacionadas aos dados de testagem utilizam a unidade de teste especificada. As unidades válidas são 'pessoas testadas' (número de pessoas testadas), 'testes realizados' (número de testes realizados, uma única pessoa pode ser testada mais de uma vez em um determinado dia) e 'amostras testadas' (número de amostras testadas. Em alguns casos, mais de uma amostra pode ser necessária para realizar um determinado teste).

Para países que não relatam dados de vacinação diariamente, assumimos que a vacinação mudou igualmente diariamente ao longo de qualquer período em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que é então calculada em uma janela móvel de 7 dias.

* `people_vaccinated_per_hundred` - Número total de pessoas que receberam pelo menos uma dose da vacina por 100 pessoas na população total.

## Análise Exploratória dos Dados

A análise exploratória de dados (AED) desempenha um papel fundamental em qualquer projeto de análise de dados por várias razões:

1. **Compreensão dos dados:** A AED ajuda a entender a natureza dos dados com os quais estamos lidando. Isso inclui a identificação de tipos de variáveis, distribuições, valores ausentes e possíveis outliers.

2. **Detecção de padrões:** Ao visualizar e resumir os dados de diferentes maneiras, podemos identificar padrões, tendências e relações entre variáveis. Isso nos ajuda a formular hipóteses e orientar análises mais avançadas.

3. **Validação de pressupostos:** Antes de aplicar modelos estatísticos ou algoritmos de machine learning, é importante verificar se os pressupostos subjacentes são atendidos. A AED permite testar esses pressupostos e tomar medidas corretivas, se necessário.

4. **Identificação de outliers e erros:** A AED ajuda a identificar valores atípicos e erros nos dados que podem distorcer as análises posteriores. Isso é crucial para garantir a qualidade dos resultados finais.

5. **Seleção de variáveis:** Ao explorar os dados, podemos identificar quais variáveis são mais relevantes para nossos objetivos de análise. Isso nos permite focar em aspectos importantes e descartar informações irrelevantes.

6. **Comunicação de resultados:** Gráficos e resumos estatísticos gerados durante a AED são úteis para comunicar insights e resultados de forma clara e acessível a diferentes partes interessadas.

Em resumo, a análise exploratória de dados é uma etapa essencial em qualquer projeto de análise de dados, pois nos ajuda a compreender, validar e extrair informações valiosas dos dados, orientando análises mais avançadas e a tomada de decisões fundamentadas.

#### Medição por 100 mil habitantes

Medir casos e mortes por 100 mil habitantes é uma prática comum na epidemiologia e na análise de dados de saúde pública. Essa medida permite uma comparação mais justa entre diferentes regiões, países ou populações, ajustando os números para considerar o tamanho da população.

Quando os casos e as mortes são expressos por 100 mil habitantes, isso significa que os números são divididos pelo tamanho da população e multiplicados por 100 mil. Isso permite uma compreensão melhor da proporção de casos ou mortes em relação ao tamanho da população, o que é útil para avaliar a gravidade de uma situação em termos de saúde pública.

Essa métrica é especialmente útil ao comparar áreas com populações muito diferentes, pois fornece uma base per capita para comparação. Por exemplo, um país com uma grande população pode ter um número absoluto maior de casos do que um país com uma população menor, mas quando os casos são ajustados por 100 mil habitantes, podemos avaliar mais claramente a extensão da propagação da doença em relação ao tamanho da população.

Agora o mesmo gráfico mas com os dados de casos e mortes por 100 mil habitantes:

```
fig, ax = plt.subplots(nrows= 2, ncols= 1, figsize=(10, 5))

df3.groupby('state')['totalCases_per_100k_inhabitants'].max().sort_values(ascending=False).plot(kind='bar', ax=ax[0])
ax[0].set_title('Casos por estado a cada 100 mil habitantes')
ax[0].set_xlabel('')
ax[0].set_ylabel('Número de casos')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=0)

df3.groupby('state')['deaths_per_100k_inhabitants'].max().sort_values(ascending=False).plot(kind='bar', ax=ax[1])
ax[1].set_title('Mortes por estado a cada 100 mil habitantes')
ax[1].set_xlabel('')
ax[1].set_ylabel('Número de mortes')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=0)

plt.tight_layout()
```

* Aqui temos um panorama mais justo sobre os casos, interessante ver que Maranhão foi o estado com menos casos e menos mortes.

#### Casos por região do Brasil

Aqui vamos verificar o número de casos por região do Brasil, verificar em qual região houve mais casos e mortes registrados.

Primeiro criarei um dicionário com os estados e suas regiões correspondentes, e depois irei usar a função `map()`

```
# Dicionário mapeando cada sigla de estado para sua região correspondente
regioes = {
    'AC': 'Norte',
    'AL': 'Nordeste',
    'AP': 'Norte',
    'AM': 'Norte',
    'BA': 'Nordeste',
    'CE': 'Nordeste',
    'DF': 'Centro-Oeste',
    'ES': 'Sudeste',
    'GO': 'Centro-Oeste',
    'MA': 'Nordeste',
    'MT': 'Centro-Oeste',
    'MS': 'Centro-Oeste',
    'MG': 'Sudeste',
    'PA': 'Norte',
    'PB': 'Nordeste',
    'PR': 'Sul',
    'PE': 'Nordeste',
    'PI': 'Nordeste',
    'RJ': 'Sudeste',
    'RN': 'Nordeste',
    'RS': 'Sul',
    'RO': 'Norte',
    'RR': 'Norte',
    'SC': 'Sul',
    'SP': 'Sudeste',
    'SE': 'Nordeste',
    'TO': 'Norte'
}

# Mapeando os estados para suas respectivas regiões usando o dicionário 'regioes'
df3['Região'] = df3['state'].map(regioes).copy()
```

```
fig, ax = plt.subplots(nrows= 2, ncols= 1, figsize=(10, 5))

df3.groupby('Região')['totalCases_per_100k_inhabitants'].sum().sort_values(ascending=False).plot(kind='bar', ax=ax[0])
ax[0].set_title('Casos por região a cada 100 mil habitantes')
ax[0].set_xlabel('')
ax[0].set_ylabel('Número de casos')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=0)

df3.groupby('Região')['deaths_per_100k_inhabitants'].sum().sort_values(ascending=False).plot(kind='bar', ax=ax[1])
ax[1].set_title('Mortes por região a cada 100 mil habitantes')
ax[1].set_xlabel('')
ax[1].set_ylabel('Número de mortes')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=0)

plt.tight_layout()
```

* Proporcionalmente, houve mais morte nas regiões norete e nordeste
